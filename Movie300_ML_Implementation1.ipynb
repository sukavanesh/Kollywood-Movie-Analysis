```python
dfxml = dfx[['Movie_name','Movie_Genre_Num','Movie_Certification_Num','Release_Month_Num','Runtime_Minutes','Movie_Critic_Rating','Movie_User_Rating']]
dfxml.head()
```

```python
#target variable
y = dfxml['Movie_User_Rating']

#input dataframe
x = dfxml[['Movie_Genre_Num','Movie_Certification_Num','Release_Month_Num','Runtime_Minutes','Movie_Critic_Rating']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x1,x2,y1,y2 = train_test_split(x,y,test_size = 0.1)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x1,y1)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
pd.DataFrame(lr.coef_, index=x.columns, columns=['myval'])
```

```python
t = np.array(dfxml.loc[8][['Movie_Genre_Num','Movie_Certification_Num','Release_Month_Num','Runtime_Minutes','Movie_Critic_Rating']])
```

```python
lr.predict([t])
```

```python
f = []
k = []
for i in range(0,300):
    b = np.array(dfxml.loc[i][['Movie_Genre_Num','Movie_Certification_Num','Release_Month_Num','Runtime_Minutes','Movie_Critic_Rating']])
    f.append(lr.predict([b]))
    k.append(np.array(dfxml.loc[i][['Movie_name']]))
```

```python
K = pd.DataFrame(k,columns=['Movie_name'])
K
```

```python
F=pd.DataFrame(f,columns=['Machine_Predicted_Rating'])
F
```

```python
J = dfxml[['Movie_Critic_Rating','Movie_User_Rating']]
```

```python
final = pd.concat([K,J,F],axis=1)
final
```
